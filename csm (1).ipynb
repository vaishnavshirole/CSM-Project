```python
import sqlite3

conn = sqlite3.connect("sales.db")
```

```python
accounts.to_sql("accounts", conn, index=False)

products.to_sql("products", conn, index=False)

pipeline.to_sql("sales_pipeline", conn, index=False)

teams.to_sql("sales_teams", conn, index=False)
```

```python
query = """
SELECT
    product,
    SUM(close_value) AS total_sales
FROM sales_pipeline
WHERE deal_stage = 'Won'
GROUP BY product
ORDER BY total_sales DESC;
"""

total_sales_by_product = pd.read_sql(query, conn)
print(total_sales_by_product)
```

```python
query = """
SELECT product,
SUM(close_value)
FROM sales_pipeline
GROUP BY product;
"""

sell_by_product = pd.read_sql(query, conn)
print(sell_by_product)
```

```python
query="""
SELECT COUNT(opportunity_id)
FROM sales_pipeline;
"""

count_of_product_id=pd.read_sql(query,conn)
print(count_of_product_id)
```

```python
query="""
SELECT SUM(close_value) AS total_revenue
FROM sales_pipeline
WHERE deal_stage='Won';
"""
total_revenue=pd.read_sql(query,conn)
print(total_revenue)
```

```python
query="""
SELECT AVG(close_value) AS average_deal
FROM sales_pipeline
WHERE deal_stage='Won';
"""
avg_deal=pd.read_sql(query,conn).round(2)
print(avg_deal)
```

```python
query="""
SELECT
product,
SUM(close_value) AS revenue

FROM sales_pipeline

WHERE deal_stage='Won'

GROUP BY product

ORDER BY revenue DESC

LIMIT 5;
"""
revenue_by_product=pd.read_sql(query,conn).round(2)
print(revenue_by_product)
```

```python
query="""
SELECT

sales_agent,
SUM(close_value) AS revenue

FROM sales_pipeline

WHERE deal_stage='Won'

GROUP BY sales_agent

ORDER BY revenue DESC LIMIT 10;
"""
revenue_by_agent=pd.read_sql(query,conn).round(2)
print(revenue_by_agent)
```

```python
query="""
SELECT

a.sector,

SUM(sp.close_value) AS revenue

FROM sales_pipeline sp

JOIN accounts a

ON sp.account=a.account

WHERE sp.deal_stage='Won'

GROUP BY a.sector

ORDER BY revenue DESC;
"""
revenue_by_sector=pd.read_sql(query,conn).round(2)
print(revenue_by_sector)
```

```python
query="""
SELECT

st.regional_office,

SUM(sp.close_value) AS revenue

FROM sales_pipeline sp

JOIN sales_teams st

ON sp.sales_agent=st.sales_agent

GROUP BY st.regional_office;
"""
revenue_by_office=pd.read_sql(query,conn).round(2)
print(revenue_by_office)
```

```python
query="""
SELECT

deal_stage,

COUNT(*) AS total

FROM sales_pipeline

GROUP BY deal_stage;
"""
count_of_deal_stage=pd.read_sql(query,conn).round(2)
print(count_of_deal_stage)
```

```python
query="""
SELECT

account,

SUM(close_value) AS revenue

FROM sales_pipeline

WHERE deal_stage='Won'

GROUP BY account

ORDER BY revenue DESC LIMIT 20;
"""
revenue_by_account=pd.read_sql(query,conn).round(2)
print(revenue_by_account)
```

```python
query="""
SELECT

strftime('%m',close_date) AS month,

SUM(close_value)

FROM sales_pipeline

WHERE deal_stage='Won'

GROUP BY month;
"""
sales_in_each_month=pd.read_sql(query,conn).round(2)
print(sales_in_each_month)
```
